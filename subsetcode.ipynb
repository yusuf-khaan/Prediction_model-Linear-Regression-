```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('Online_Retail.xlsx')
```

```python
data.drop(columns={'Country','UnitPrice','Description'})
```

```python
dic = {}
da = data['InvoiceDate']
df = pd.DataFrame(da)
df=df['InvoiceDate'].astype('str')
daaa = list(df.items())
dafa = pd.DataFrame(daaa)
dafa.drop(columns={0},inplace=True)
```

```python
dafa['Year'] = dafa['InvoiceD'].astype(str)
dafa.head()
```
